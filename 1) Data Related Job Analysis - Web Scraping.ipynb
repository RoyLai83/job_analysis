Getting details from the job sites. I have split this up into 5 section for each of the keywords, and appending the inforamtion to the same list. I can probably do a loop but wanted to seperate it as it took a long time and was worried about the computer crashing or not finishing properly.

None-the-less - I could have just put all the job listings scraped earlier into one long list and run it in one go.

```python
data = []
for job in data_scientist_links:
    url = 'https://www.seek.com.au'+job
    response = requests.get(url)
    if response.status_code == 200:
        html = response.text
        soup = BeautifulSoup(html,'lxml')
        row = {}

## scraping and putting the information into a dictionary. Using Try/Except in case there are inconsistencies with the information entered into the job advertisement (missing info will return a 'NA')
        try:
            row['job_description'] = soup.find('div',{"data-automation":"jobDescription"}).find('div',{"class":"templatetext"}).text
        except:
            try:
                row['job_description'] = soup.find('div',{"data-automation":"jobDescription"}).find('div',{"data-automation":"mobileTemplate"}).text
            except:
                row['job_description'] = 'NA'

        try:
            row['job_title'] = soup.find(name='span',attrs={"data-automation":"job-detail-title"}).text
        except:
            row['job_title'] = 'NA'
        try:
            row['advertiser'] = soup.find(name='span',attrs={"class":"_3FrNV7v _2QG7TNq E6m4BZb"}).findChild('span').text
        except:
            row['advertiser'] = 'NA'
            
        aside = soup.find('section',{"aria-labelledby":"jobInfoHeader"}).find('dl') #breaking the beautiful soup code a little bit to make it easier.
        try:
            row['posting_date'] = aside.find('dd').find(name='span',attrs={"class":"lwHBT6d"}).text
        except:
            row['posting_date'] = 'NA'
        try:
            row['location'] = aside.find(name='strong',attrs={"class":"lwHBT6d"}).text
        except:
            row['location'] = 'NA'
        try:
            row['sub_location'] = aside.find(name='span',attrs={"class":"eBOHjGN"}).text
        except:
            row['sub_location'] = 'NA'
        try:
            row['salary'] = aside.find('div').find(name='span',attrs={"class":"lwHBT6d"}).text
        except:
            row['salary'] = 'NA'
        try:
            row['job_type'] = aside.find('dd',{'data-automation':'job-detail-work-type'}).find(name='span',attrs={"class":"lwHBT6d"}).text
        except:
            row['job_type'] = 'NA'
        try:
            row['industry'] = aside.find_all(name='strong',attrs={"class":"lwHBT6d"})[-1].text
        except:
            row['industry'] = 'NA'
        try:
            row['job_category'] = aside.find_all(name='span',attrs={"class":"eBOHjGN"})[-1].text
        except:
            row['job_category'] = 'NA'
        row['search_word'] = 'data scientist'    
        row['link'] = url
        
        data.append(row)
```
